```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the dataframe
data = {
    "Month": list(range(1, 13)),
    "ESI": [929, 733, 760, 983, 733, 634, 774, 673, 787, 933, 849, 886], #available weeks in the 1 week ESI data sm_anomaly_by_station_and_esi.csv
    "2inANOM": [886, 701, 721, 934, 692, 610, 744, 623, 738, 878, 796, 831],
    "4inANOM": [841, 673, 691, 896, 665, 584, 712, 608, 702, 834, 765, 798],
    "8inANOM": [875, 693, 708, 911, 681, 595, 730, 626, 726, 872, 798, 825],
    "20inANOM": [847, 680, 696, 892, 672, 592, 719, 622, 733, 871, 773, 800],
    "40inANOM": [789, 634, 639, 831, 625, 543, 671, 567, 673, 787, 710, 736]
}

df = pd.DataFrame(data)

# Convert anomalies to percentages of ESI
for col in df.columns[2:]:  # Exclude 'Month' and 'ESI'
    df[col] = (df[col] / df["ESI"]) * 100

# Plot
plt.figure(figsize=(10, 6))

for col in df.columns[2:]:
    plt.plot(df["Month"], df[col], marker='o', label=col)

plt.xlabel("Month")
plt.ylabel("% of Total")
plt.title("Available Pairwise Soil Moisture vs ESI Observations by Depth")
plt.xticks(df["Month"])
plt.legend()
plt.grid(True)

plt.savefig("soil_moisture_anomalies.png", dpi=300, bbox_inches="tight")
# Show the plot
plt.show()
```
