In order to compare ESI to Soil Moisture Values, we need an anomaly of soil moisture values to compare apples to apples. 

In order to do this, we will use a list of functions to convert the volumetric soil moisture readings to effective saturation (which normalizes the value between 0-1 where 1 is 100% wet), and create the anomaly. 


List of functions from effective_SM_conversion_all.py module and what they do:
* SOIL_TYPE( ) -  

    Takes in a test dataframe containing volumetric soil moisture information 
    by station and date and outputs soil parameters column. 
    
    
* CALCULATE_ESM( ) -  

    converts volumetric soil moisture columns to van ganuchin effective soil 
    moisture for each level (2in, 4in, 8in, 20in, 40in) 



* ESM_ANOM( ) - 

    converts daily effective soil moisture means to rolling seven day means. Then
    subtracts this mean from the rolling mean from the all day averages to get an 
    anomaly.



* Merge( ) - 

    Merges the ESI data with SCAN at appropriate weeks. 
    
    
    
* MAKE_DF( ) - 

    Combindes all of the functions above on the TEST data to create a final dataframe
    for analysis.

# An Observation...
You will see above that we have reduced the dataframe from 40K daily values for the 6 combined stations to 3991. This is because we are reducing the original soil moisture dataset to weekly values and droping rows where there are nulls. Now, lets look at a plot of anomalies versus ESI. 

```python
#segment the dataframe to one station during one drought year
anomalies = new_df[['StationTriplet', 'ESI', 'ANOM_2in', 'ANOM_4in', 'ANOM_8in', 'ANOM_20in', 'ANOM_40in']]
one_scan = anomalies[(anomalies['StationTriplet'] == '2053:AL:SCAN') & (anomalies.index.year <= 2007) & (anomalies.index.year >= 2005)]
one_scan = one_scan[['ESI', 'ANOM_2in', 'ANOM_4in', 'ANOM_8in', 'ANOM_20in', 'ANOM_40in']]

#amplify the normalization curves to make them easier to read
one_scan['ANOM_2in'] = one_scan['ANOM_2in']*4
one_scan['ANOM_4in'] = one_scan['ANOM_4in']*4
one_scan['ANOM_8in'] = one_scan['ANOM_8in']*4
one_scan['ANOM_20in'] = one_scan['ANOM_20in']*4
one_scan['ANOM_40in'] = one_scan['ANOM_40in']*4

fig, ax = plt.subplots(figsize=(12,10))
ax.plot(one_scan['ESI'], label='ESI')
ax.plot(one_scan['ANOM_2in'], label='ANOM_2in')
ax.plot(one_scan['ANOM_4in'], label='ANOM_4in')
ax.plot(one_scan['ANOM_8in'], label='ANOM_8in')
#ax.plot(one_scan['ANOM_20in'], label='ANOM_20in')
#ax.plot(one_scan['ANOM_40in'], label='ANOM_40in')
ax.set_title('Drought Year ALEXI ESI vs USDA Site 2053:AL:SCAN Effective Soil Moisture Anomalies by Depth')
ax.set_ylabel('Anomaly Value')
ax.set_xlabel('Date')
ax.legend()
```
